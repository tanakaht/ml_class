```python
import numpy as np

np.random.seed(seed=32)
n = 200;
X = 3 * (np.random.rand(n, 4) - 0.5)
W = np.array([[2, -1, 0.5],
                      [-3, 2, 1],
                      [1, 2, 3]])
y = np.argmax( np.concatenate((X[:, 1:3], np.ones((n, 1))), axis=1) @ W + 0.5 * np.random.randn(n, 3) , axis=1)
datasetv = {'data': X, 'target': y}
print(X.shape)
plt.scatter(X[:, 1], X[:, 2], c=y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


class LogisticRegression:
    def __init__(self, X, y, lam=1, w0=None):
        self.X = np.concatenate((X, np.ones((n, 1))), axis=1) # classifier adaptable to offset
        self.y = y
        self.lam = lam
        if w0 is not None:
            self.w = w0
        else:
            self.w = np.ones(self.X.shape[1])

    def J(self):
        p = 1/(1 + np.exp(-self.y * (self.X@self.w)))
        return - np.sum(np.log(p)) + self.lam * self.w@self.w

    def grad_J(self):
        p = 1/(1 + np.exp(-self.y * (self.X@self.w)))
        return - np.sum(self.X.T * self.y * (1-p), axis=1) + 2 * self.lam * self.w

    def hess_J(self):
        p = 1/(1 + np.exp(-self.y * (self.X@self.w)))
        return (p * (1-p) * self.X.T)@self.X + 2 * self.lam 
    
    def optimize_with_sg(self, learning_rate=0.01, max_itr=1000, eps=1e-6):
        j_history = []
        pre = np.inf
        for itr in range(0, max_itr):
            if itr % 1 == 0:
                now = self.J()
                j_history.append((itr, now))
                if(abs(pre - now) < eps):
                    print(f"iteration stop at {itr}")
                    break
                else:
                    pre = now
            self.w = self.w - learning_rate * self.grad_J()
        return np.array(j_history)
        
        
    def optimize_with_newton(self, max_itr=1000, eps=1e-6):
        j_history = []
        pre = np.inf
        for itr in range(0, max_itr):
            if itr % 1 == 0:
                now = self.J()
                j_history.append((itr, now))
                if(abs(pre - now) < eps):
                    print(f"iteration stop at {itr}")
                    break
                else:
                    pre = now
            self.w = self.w - np.linalg.solve(self.hess_J(), self.grad_J())
        return np.array(j_history)

X = datasetiv['data']
y = datasetiv['target']

lr_sg = LogisticRegression(X, y, lam=1)
j_history_sg = lr_sg.optimize_with_sg()
print(lr_sg.w)
print(j_history_sg[-1])

lr_newton = LogisticRegression(X, y, lam=1)
j_history_newton = lr_newton.optimize_with_newton()
print(lr_newton.w)
print(j_history_newton)

plt.title("J(w)")
plt.xlabel("iteration")
plt.ylabel("J(w)")
plt.yscale("log")
plt.plot(j_history_sg[:, 0], j_history_sg[:, 1], label="steepest gradient method1")
plt.plot(j_history_newton[:, 0], j_history_newton[:, 1], label="Newton")
plt.legend()
plt.show()

plt.title("J(w)-J(w_hat)")
plt.xlabel("iteration")
plt.ylabel("J(w)-J(w_hat)")
plt.yscale("log")
plt.plot(j_history_sg[:, 0], j_history_sg[:, 1] - j_history_sg[-1, 1], label="steepest gradient method")
plt.plot(j_history_newton[:, 0], j_history_newton[:, 1] - j_history_newton[-1, 1], label="Newton")
plt.legend()
plt.show()

plt.title("steepest gradient method")
plt.xlabel("2nd element")
plt.ylabel("3rd element")
plt.scatter(X[:, 1], X[:, 2], c=y)
plt.plot(X[:, 1], -(X[:, 1]*lr_sg.w[1] + lr_sg.w[4])/lr_sg.w[2], label="wtx=0")
plt.plot(X[:, 1], -(X[:, 1]*lr_sg.w[1] + lr_sg.w[4]+1)/lr_sg.w[2], label="wtx=1")
plt.plot(X[:, 1], -(X[:, 1]*lr_sg.w[1] + lr_sg.w[4]-1)/lr_sg.w[2], label="wtx=-1")
plt.legend()
plt.show()

plt.title("newton")
plt.xlabel("2nd element")
plt.ylabel("3rd element")
plt.scatter(X[:, 1], X[:, 2], c=y)
plt.plot(X[:, 1], -(X[:, 1]*lr_newton.w[1] + lr_newton.w[4])/lr_newton.w[2], label="wtx=0")
plt.plot(X[:, 1], -(X[:, 1]*lr_newton.w[1] + lr_newton.w[4]+1)/lr_newton.w[2], label="wtx=1")
plt.plot(X[:, 1], -(X[:, 1]*lr_newton.w[1] + lr_newton.w[4]-1)/lr_newton.w[2], label="wtx=-1")
plt.legend()
plt.show()
```
